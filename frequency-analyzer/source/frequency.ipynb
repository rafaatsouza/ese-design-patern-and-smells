```python
import matplotlib.pyplot as plt
import numpy as np

from analyzer import Analyzer

analyzer = Analyzer('../design_patterns_smells.csv')
analyzer.run()

smells = ['God Class', 'Refused Bequest', 'Feature Envy', 'Long Method']
design_pattern_colors = ['black', '#2e2e2e',
                         '#3d3d3d', '#545454', '#707070', '#878787']
design_patterns = ['Bridge', 'Composite', 'Chain of Responsibility',
                   'Template Method', 'Factory Method', 'Prototype']
design_pattern_labels = ['Bridge', 'Composite', 'Chain of R.',
                          'Template M.', 'Factory M.', 'Prototype']

summary = analyzer.summary
y_pos = np.arange(len(design_patterns))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 3))

frequencies = []
for i in range(len(design_patterns)):
    frequencies.append(
        float(summary['patterns'][design_patterns[i]]['smelly_classes']/summary['patterns'][design_patterns[i]]['classes']))

ax.barh(y_pos, frequencies, align='center', color=design_pattern_colors, ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(design_pattern_labels) 
ax.invert_yaxis()
ax.set_xlim([0, 1])
ax.set_xlabel('Frequency')
ax.set_title('Bad Smells frequency by design patterns')
plt.grid()

plt.show()
```

```python
fig, ax = plt.subplots(int(len(smells)/2), 2, figsize=(13, 5))

for i in range(len(smells)):
    ax = plt.subplot(int(len(smells)/2), 2, (i+1))
    frequencies = []
    for j in range(len(design_patterns)):
        frequencies.append(
            float(summary['patterns'][design_patterns[j]][smells[i]]['freq_in_all']))

    ax.barh(y_pos, frequencies, align='center', color=design_pattern_colors, ecolor='black')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(design_pattern_labels) 
    ax.invert_yaxis()
    ax.set_xlim([0, 1])
    ax.set_xlabel('Frequency')
    ax.set_title('{} frequency across all classes'.format(smells[i]))
    plt.grid()

plt.tight_layout()
plt.show()
```
