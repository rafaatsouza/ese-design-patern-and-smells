```python
smells = ['God Class', 'Refused Bequest', 'Feature Envy', 'Long Method']

current = 1
plt.subplots(len(smells), 2, figsize=(20, 13))

for i in range(len(smells)):
    plt.subplot(len(smells), 2, current)
    frequencies = []
    for dp in design_patterns:
        frequencies.append(
            float("%.3f" % summary[dp][smells[i]]['freq_in_all']))

    plt.bar(y_pos, frequencies, align='center', alpha=0.5)
    plt.xticks(y_pos, design_patterns)
    plt.ylabel('Frequency')
    plt.title('{} frequency across all classes'.format(smells[i]))
    plt.grid()
    plt.ylim([0,1.1])
    current = current + 1

    plt.subplot(len(smells), 2, current)
    frequencies = []
    for dp in design_patterns:
        frequencies.append(
            float("%.3f" % summary[dp][smells[i]]['freq_in_smelly']))

    plt.bar(y_pos, frequencies, align='center', alpha=0.5, )
    plt.xticks(y_pos, design_patterns)
    plt.ylabel('Frequency')
    plt.title('{} frequency across smelly classes'.format(smells[i]))
    plt.grid()
    plt.ylim([0,1.1])
    current = current + 1

plt.tight_layout()
plt.show()
```
